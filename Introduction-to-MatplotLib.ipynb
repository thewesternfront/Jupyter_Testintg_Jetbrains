## MatplotLib Example Workflow

```python
# 1 Import MaoplotLib
import matplotlib.pyplot as plt
# 2 prepare data x=[], y=[]
# 3 Set up the plot
fig, ax = plt.subplots(figsize = (10, 5))  # width, height, this is the aspect ratio, 5,10 is portrait, 10,5 is landscape
# 4 plot data
ax.plot(x, y)
# 5 Customize Plot
ax.set(title = "simple plot", xlabel = "x axis", ylabel = "y axis")
# 6 Save & Show
fig.savefig("sampleplot.png")
```

```python
# Create a plot from a Dictionary
nut_butter_prices = {'peanut': 10, 'cashew': 14, 'almond': 8}
fig, ax = plt.subplots()
# HERE: we need ot take the un-hashable dict values and typecast them to a LIST
key = (list(nut_butter_prices.keys()))
val = (list(nut_butter_prices.values()))
ax.barh(key, val)
ax.set(title = "Troy's Nut Butter Store", ylabel = "Butter Type", xlabel = "Price ($USD)");
```

## Plotting from Pandas DataFrames

```python
import pandas as pd
```

```python
# Make a DataFrame
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020', periods=1000))
print(ts)
# get the cumulative sum
ts = ts.cumsum()
print(ts)
ts.plot()
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$\,\.]', '', regex=True)
car_sales
```

```python
# Remove the extraneous zeros for the cents
# first lets look at the data types
type(car_sales["Price"])
```

```python
# Noew we see what types the array element is (string)
type(car_sales["Price"][0])
```

```python
# Now we will truncate it using the str() method
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
# notice the usage of astype() to convert th string to an int so that it does not jsut concatenate the strings, but actually cumulatively represents the integer value of the Price
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plotting the total sales
# This is the pandas wrap up of the matplotlib API
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# create a scatter plot for the Odometer and Price
# convert the price to an int so it can be plotted
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer", y="Price", kind="scatter");
```
