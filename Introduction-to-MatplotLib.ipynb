## MatplotLib Example Workflow

```python
# 1 Import MaoplotLib
import matplotlib.pyplot as plt
# 2 prepare data x=[], y=[]
# 3 Set up the plot
fig, ax = plt.subplots(figsize = (10, 5))  # width, height, this is the aspect ratio, 5,10 is portrait, 10,5 is landscape
# 4 plot data
ax.plot(x, y)
# 5 Customize Plot
ax.set(title = "simple plot", xlabel = "x axis", ylabel = "y axis")
# 6 Save & Show
fig.savefig("sampleplot.png")
```

```python
# Create a plot from a Dictionary
nut_butter_prices = {'peanut': 10, 'cashew': 14, 'almond': 8}
fig, ax = plt.subplots()
#ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
key = nut_butter_prices.keys()
val =  nut_butter_prices.values()
ax.bar(key, val)
ax.set(title = "Troy's Nut Butter Store", ylabel = "Price ($USD)");
```

```python
# Create a plot from a Dictionary
nut_butter_prices = {'peanut': 10, 'cashew': 14, 'almond': 8}
fig, ax = plt.subplots()
# HERE: we need ot take the un-hashable dict values and typecast them to a LIST
key = (list(nut_butter_prices.keys()))
val = (list(nut_butter_prices.values()))
ax.barh(key, val)
ax.set(title = "Troy's Nut Butter Store", ylabel = "Butter Type", xlabel = "Price ($USD)");
```
