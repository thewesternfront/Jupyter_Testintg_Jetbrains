## MatplotLib Example Workflow

```python
# 1 Import MaoplotLib
import matplotlib.pyplot as plt
# 2 prepare data x=[], y=[]
# 3 Set up the plot
fig, ax = plt.subplots(figsize = (10, 5))  # width, height, this is the aspect ratio, 5,10 is portrait, 10,5 is landscape
# 4 plot data
ax.plot(x, y)
# 5 Customize Plot
ax.set(title = "simple plot", xlabel = "x axis", ylabel = "y axis")
# 6 Save & Show
fig.savefig("sampleplot.png")
```

```python
# Create a plot from a Dictionary
nut_butter_prices = {'peanut': 10, 'cashew': 14, 'almond': 8}
fig, ax = plt.subplots()
# HERE: we need ot take the un-hashable dict values and typecast them to a LIST
key = (list(nut_butter_prices.keys()))
val = (list(nut_butter_prices.values()))
ax.barh(key, val)
ax.set(title = "Troy's Nut Butter Store", ylabel = "Butter Type", xlabel = "Price ($USD)");
```

```python
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$\,\.]', '', regex=True)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
# notice the usage of astype() to convert th string to an int so that it does not jsut concatenate the strings, but actually cumulatively represents the integer value of the Price
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

### Which method should be used? (pyplot vs matplotlib OO method?)

* When plotting something quickly, OK to use the pyplot method
* When plotting something more advenced, use the matplotlib OO method

```python
# Explore patients over the age of 50
over_50 = heart_disease[heart_disease["age"] > 50]
heart_disease.head()
```

```python
over_50.head()
```

```python
# Scatter plt of age and cholesterol
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

```python
# Matplotlib method now instead of plot()
fig, ax = plt.subplots(figsize=(10,6))
# ax.set_xlim([45, 100])
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])


plt.axhline(y=over_50['chol'].mean(), color='r', linestyle='-')
plt.axvline(x=over_50['age'].mean(), color='g', linestyle='-')

ax.set(title = "Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol Level")
ax.legend(*scatter.legend_elements(), title="Target")

# plt.legend(over_50['age'].mean())
# plt.legend(over_50['chol'].mean())
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Chol")
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal and vertical line
ax.axhline(over_50["chol"].mean(), color='g', linestyle='--')
```

```python
# Adding a Subplot - Chol, Age, Thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1)

# Add data to axis 0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
# Customize ax0 subplot
ax0.set(title="Cholesterol Levels and Heart Diesase", xlabel = "Age", ylabel = "Cholesterol Level")
# Add a Legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a Horizontal Line
ax0.axhline(y=over_50["chol"].mean(), color='g', linestyle='--')


# Repeat same but replace chol with thalach
# Add data to axis 0
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
# Customize ax0 subplot
ax1.set(title="Thalach Values and Heart Diesase", xlabel = "Age", ylabel = "Max Heart Rate after exercise")
# Add a Legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a Horizontal Line
ax1.axhline(y=over_50["thalach"].mean(), color='r', linestyle='--')
ax1.axhline(y=over_50["thalach"].std(), color='g', linestyle='--')

# Add a title for the entire plot
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight='bold');
```
