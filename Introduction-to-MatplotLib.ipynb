## MatplotLib Example Workflow

```python
# 1 Import MaoplotLib
import matplotlib.pyplot as plt
# 2 prepare data x=[], y=[]
# 3 Set up the plot
fig, ax = plt.subplots(figsize = (10, 5))  # width, height, this is the aspect ratio, 5,10 is portrait, 10,5 is landscape
# 4 plot data
ax.plot(x, y)
# 5 Customize Plot
ax.set(title = "simple plot", xlabel = "x axis", ylabel = "y axis")
# 6 Save & Show
fig.savefig("sampleplot.png")
```

## Create some plots:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
# Create some data
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
print(x[:10])
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

## Use same data to make a Scatter Plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Create another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Create a plot from a Dictionary
nut_butter_prices = {'peanut': 10, 'cashew': 14, 'almond': 8}
fig, ax = plt.subplots()
#ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
key = nut_butter_prices.keys()
val =  nut_butter_prices.values()
ax.bar(key, val)
ax.set(title = "Troy's Nut Butter Store", ylabel = "Price ($USD)");
```

```python
fig, ax = plt.subplots()
ax.barh(key, val)
plt.show()
```

```python
# Histograms
xd = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(xd)
plt.show()
```

## Two options for SubPlots


```python
key = [1, 2, 3, 4, 5]
val = [11, 22, 33, 44, 55]
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
plt.show(ax1.bar(key, val))
```
